```julia
"""
    walk(t) -> trajectory

模拟随机行走
"""

function walk(T)
    x = 0 #初始位置
    trajectory = [x] #初始化轨迹
    
    for t in 1:T
        if rand() < 0.5
            x += 1
        else
            x -= 1
        end
        push!(trajectory, x)
    end
    trajectory
end
    
```

```julia
?walk
```

```julia
using GR, Interact #还有Plots 效率慢

T = 100
plot(0:T, walk(100))
```

```julia
@manipulate for T in 10:100
    plot(0:T, walk(T))
end
```

```julia
using JLD2

traj = walk(1000)

@save "data.jld2" traj
```

```julia
@load "data.jld2" traj
typeof(traj)
```
